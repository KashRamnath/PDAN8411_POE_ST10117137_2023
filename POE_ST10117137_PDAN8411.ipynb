# What Is Text Processing?

The definition of text processing is the computational process and techniques of analysing and manipulation of textual information through transforming raw textual data into a form that is easy to analyse, read and extract meaningful data. This is also known as the automatic transmission of information in a sequential approach, Technopedia (2023).

'Text processing has a direct application to Natural Language Processing, also known as NLP. NLP is aimed at processing the languages spoken or written by humans when they communicate with one another', Tutorialspoint (2023).

'Machine learning (ML) for natural language processing (NLP) and text analytics involves using machine learning algorithms and “narrow” artificial intelligence (AI) to understand the meaning of text documents. These documents can be just about anything that contains text: social media comments, online reviews, survey responses, even financial, medical, legal and regulatory documents. In essence, the role of machine learning and AI in natural language processing and text analytics is to improve, accelerate and automate the underlying text analytics functions and NLP features that turn this unstructured text into useable data and insights', lexalytics (2022).

The process of performing this may include any of the following techniques (These are just a few popular techniques, although there are more):

- Tokenization
- Removing stopworda
- Normalization
- POS tagging
- NER
- Dependency Parsing
- Sentiment analysis
- Text classiification

```python
# the above mentioned variables to be treated for null removal
feature_lst = ["company_profile", "description", "requirements"]

for col in feature_lst:
    # null values change to none if fraudulent is 0
    df.loc[(df[col].isnull()) & (df["fraudulent"] == 0), col] = "none"
    
    # null values change to missing if fraudulent is 1
    df.loc[(df[col].isnull()) & (df["fraudulent"] == 1), col] = "missing"
```

```python
# create cloumns of lengths above mentioned variables company_profile, decription and requirements
for num,col in enumerate(feature_lst):
    df[str(num)] = df[col].apply(len)

# name the columns
df = df.rename({"0": "profile_length", "1": "description_length", "2": "requirements_length"}, axis=1)
```

# ML Model

```python
# display columns
df.columns
```

```python
# we use company profile as the determinable factor if a job posting 
# is fraudulent or not, as shown above, is the most suitable for 
# detection of fake or real job postings
X1_profile = df["company_profile"]

# the y variable, being fraudulent or not
y1 = df["fraudulent"]

# we split into 20% test, 80% train, with a fixed random_state 
# with the purpose of keeping results temporarily consistent
# while we test validation - random state can be changed afterwards
X1_profile_train, X1_profile_test, y1_train, y1_test = train_test_split(X1_profile, y1, test_size=0.2, random_state=42)
```

```python
# we create a function for removing punctuation and stopwords
def text_process(text):

    # take out punctuations
    noPunctuation = [char for char in text if char not in string.punctuation]
    
    # combine '' to form strings 
    noPunctuation = "".join(noPunctuation)
    
    # deplete the STOPWORDS
    return [word for word in noPunctuation.split() if word.lower() not in stopwords.words("english")]
```

```python
# creation of the pipeline
my_pipeline = Pipeline([("bow no func", CountVectorizer()),
                       ("NB_classifier", MultinomialNB())])
```

```python
# fit the pipleine with training data
my_pipeline.fit(X1_profile_train, y1_train)
```

```python
# pedict the test data
NB_pred = my_pipeline.predict(X1_profile_test)
```

```python
# now we see the results of the prediction
# via the classificationb_report method
print(classification_report(y1_test, NB_pred))
```

```python
# create fusion matrix
conf_mat = confusion_matrix(y1_test, NB_pred)

# visualise the matrix
f, ax = plt.subplots(figsize=(5,5))
sea.heatmap(conf_mat, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()

 # print basic matrix as well
print(confusion_matrix(y1_test, NB_pred))
```

```python
# use pipline
# analyzer: callable. A function to handle preprocessing, tokenization and n-grams generation.
NB_func_tfidf_pipeline = Pipeline([("bow with func", CountVectorizer(analyzer=text_process)),
                              ("tfidf", TfidfTransformer()),
                              ("NB_classifier", MultinomialNB())])
```

```python
# set x variable
X_profile = df["company_profile"]

#set y variable
y = df["fraudulent"]

# train dataset 80% train, 20% test
X_profile_train, X_profile_test, y_train, y_test = train_test_split(X_profile, y, test_size=0.2, random_state=42)
```

```python
# fit pipeline
NB_func_tfidf_pipeline.fit(X_profile_train, y_train)
```

```python
# predict test data
NB_func_tfidf_pred = NB_func_tfidf_pipeline.predict(X_profile_test)

# print predicted values
print(NB_func_tfidf_pred)
```

```python
# evaluate performance of the model
print(classification_report(y_test, NB_func_tfidf_pred))

# we can see that the model performs very well
# with pipleine functioning correctly
```

```python
conf_mat = confusion_matrix(y_test, NB_func_tfidf_pred)

# print confusion matrix to visualise result performance
f, ax = plt.subplots(figsize=(5,5))
sea.heatmap(conf_mat, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show()

# also print regular matrix
print(confusion_matrix(y_test, NB_func_tfidf_pred))

# the model predicts/classifies most values correctly
```
